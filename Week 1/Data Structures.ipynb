```python
#Dictionary: An unordered collection key-value pairs item created in a curly brackets or with key word dict before a normal bracket.
MY_Dictionary = dict(name = 'Lola', Age = 20, City='Nairobi') #1st way to create a dictionary
MY_Dictionary
```

```python
My_Dictionary = dict([('name','Lola'), ('Age', 20), ('City','Nairobi')]) # 2nd way to create a dictionary
My_Dictionary
```

```python
My_Dictionary = {'name' : 'Lola', 'Age': 20, 'City': 'Nairobi'} #3rd way to create a dictionary
My_Dictionary
```

```python
My_Dictionary = {'name' : 'Lola', 'Age': 20, 'City': 'Nairobi'} #Accessing a value in a dictionary
Age= My_Dictionary.get('Age')
Age
```
