Python uses two's complement for signed integers.
### Step 1: Binary of 10;   10 = 1010
Conceptually, positive numbers have infinite leading zeros:  
...00000000001010
### Step 2: Flip all bits
   ...11111111110101
### Step 3: Understand Two's Complement
In two's complemnt:
Mathematical Rule: x = -(x+1) -- 10 = -(10+1)
To find the value of a negative number:
1. Flip bits
2. Add 1


Bitwise Negation

```python
print(~x)  # Output: -11
```

```python
#(right shift): Each shift divides by 2 (floor division for integers)
# Effect: Move all bits n positions to the right.
# Think of right shift as “dropping the last n bits”.

print(x >> 2)  # Output: 2
10 = 1010
101 #first shift dropped the last bit which is 0
10  #second shift dropped the last bit which is 1
10/2 # fisrt shift diivion by 2
5/2  # second shift division by 2
2 # final outcome is 2 because we're doing a floor division
```

```python
print(x>>3) # output is 1, basically we're moving to the right 3 times
10/2
5/2
2/2
1
```

```python
#(left shift): Each shift multiplies by 2.
# Effect: Move all bits n positions to the left.
# Think of left shift as “adding zeros to the right” like multiplying by 10 in decimal, but in base 2.

print(x << 3)  # Output: 80
10*2
20*2
40*2
80

```

```python
#binary explanation 
10 = 1010
1010*10 #first shift to the left
10100*10 # second shift to the left
101000*10 #third shift to the left
1010000
```

```python
print(x<<5)
10*2
20*2
40*2
80*2
160*2
#final outcome 320
```

Membership Operators
Python has two membership operators
    1. in
    2. not in
They check whether a value exists in a sequence or collection.
Sequences include:
    -strings
    -lists
    -tuples
    -sets
    -dictionaries(keys)
    


```python
a=[1,2,3,4,5]
```

```python
print(2 in a)
```

```python
print(8 in a)
```

```python
print(4 not in a)
```

```python
print(10 not in a )
```

Identity Operators
Python has two identity operators:
    1. is
    2. is not
They check if two objects are the same object in memory, not just equal in value.


```python
a= [1,2,3,4,5]
b= [1,2,3,4,5]
```

```python
print(a is b)
```
